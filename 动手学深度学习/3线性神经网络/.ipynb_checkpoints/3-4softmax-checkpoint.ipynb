##### softmax函数能够将未规范化的预测变换为非负数并且总和为1，同时让模型保持 可导的性质。 
##### 为了完成这一目标，我们首先对每个未规范化的预测求幂，这样可以确保输出非负。 为了确保最终输出的概率值总和为1，我们再让每个求幂后的结果除以它们的总和。
