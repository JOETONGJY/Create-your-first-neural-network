```python
import torch
a = torch.arange(3).reshape((3, 1))
b = torch.arange(2).reshape((1, 2))
a, b
```

##### 如果让ab相加，它们的形状不匹配。 我们将两个矩阵广播为一个更大的矩阵，如下所示：矩阵a将复制列， 矩阵b将复制行，然后再按元素相加

```python
X = a + b
X
```

##### 张量中的元素可以通过索引访问

```python
X[-1], X[0:2]
```

##### 除读取外，我们还可以通过指定索引来将元素写入矩阵

```python
X[1, 1] = 9
X
```
