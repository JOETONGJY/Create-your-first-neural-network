##### 跳元模型 skip-gram

```python
# 跳元模型假设一个词可以用来在文本序列中生成其周围的单词。
# 上下文词是在给定中心词的情况下独立生成的（即条件独立性）
# 所以每个词都通过两个d维向量表示，分别表示vi和ui，也就是作为中心词和上下文词的向量
# 在这里 v = input vector, u = output vector or context vector 纯约定俗成
# 给定中心词wc(c为词典索引),生成任何上下文词wo的条件概率可以对点积建模:
#  P(wo | wc) = exp(uoTvc) / ∑ i∈V exp(uiTvc)
# 其中 V是词表索引集
```

##### 连续词袋 CBOW continuous bags of word

```python
# 根据上下文作为条件判断中心词生成的概率
# skip-gram根据中心词vi判断context生成的概率
# 在CBOW中, vi表示上下文词,ui表示中心词向量
```

```python
# 区别
# 1.CBOW训练块，skip-gram慢
# 2.CBOW有更多的向量细节
# 3.对于低频词，skip-gram更好

# 你可以意识到，skipgram和CBOW,解决的是：词 → 向量
# 现在我们解决的更多是 某一个语境中的词 → 向量
# 一个词，只有一个向量，是不合适的
# 现在做的是 contextual embedding
```
